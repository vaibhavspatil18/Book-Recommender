```python
# popularity based recommendation system. displaying top 50 books having highest avg voting (at least 200 votings)
```

```python
books_with_ratings = ratings.merge(books, on='ISBN') #merging ratings and books dataset
```

```python
books_with_ratings.shape
```

```python
num_rating_df = books_with_ratings.groupby('Book-Title').count()["Book-Rating"].reset_index()
num_rating_df.rename(columns = {"Book-Rating":"Num_Ratings"},inplace=True)
num_rating_df
```

```python
avg_rating_df = books_with_ratings.groupby('Book-Title').mean()["Book-Rating"].reset_index()
avg_rating_df.rename(columns = {"Book-Rating":"Avg_Ratings"},inplace=True)
avg_rating_df
```

```python
avg_rating_df['Avg_Ratings'] = np.round(avg_rating_df['Avg_Ratings'], decimals = 3)
avg_rating_df
```

```python
popularity_df = num_rating_df.merge(avg_rating_df, on = "Book-Title")
popularity_df
```

```python
popularity_df = popularity_df[popularity_df['Num_Ratings']>= 200].sort_values("Avg_Ratings", ascending = False).head(50)
popularity_df
```

```python
popularity_df = popularity_df.merge(books, on="Book-Title").drop_duplicates("Book-Title")[["Book-Title", "Book-Author", "Image-URL-M", "Year-Of-Publication", "Num_Ratings","Avg_Ratings"]]
```

```python
popularity_df
```

```python
# collaborative filtering based recommendation
```

```python
books_with_ratings
```

```python
x = books_with_ratings.groupby('User-ID').count()['Book-Rating'] > 200
Users_eligible = x[x].index
```

```python
filtered_rating = books_with_ratings[books_with_ratings["User-ID"].isin(Users_eligible)]
filtered_rating.head(5)
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating["Book-Title"].isin(famous_books)]
```

```python
final_ratings.head(5)
```

```python
final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index = "Book-Title", columns = "User-ID",values="Book-Rating")
```

```python
pt
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
cosine_similarity(pt).shape
```

```python
similarity_score
```

```python
similarity_score[0]
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key = lambda x:x[1], reverse=True)[1:6]
    
    data = []
    for i in similar_items:
        item =[]
        temp_df = books[books["Book-Title"]==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Title"].values))
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Author"].values))
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Image-URL-M"].values))
        
        data.append(item)
    return data
```

```python
recommend("Message in a Bottle")
```
